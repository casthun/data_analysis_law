```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# 리스트 하나만 들어가는 경우, 리스트는 y값으로 간주, x값은 [0,1,2,3]

fig, ax = plt.subplots()

# fig: 전체 그림(캔버스)에 해당함.
"""
  하나 이상의 Axes(그래프)를 포함할 수 있음.
  전체적인 크기, 저장, 레이아웃 등을 제어할 때 사용
"""

# ax: 실제 그래프를 그리는 축(Axes) 객체
"""
실제로 데이터를 시각화하는 영역임.
x축, y축, 눈금, 제목, 라벨 등을 조절 가능
선그래프, 산점도, 막대그래프 등을 그릴 때 사용
"""

ax.plot([1,2,3,4])

plt.show()
```

# Bar plot

```python
import pandas as pd

df = pd.read_csv("pandas_sample#1.csv", encoding='cp949')
df["시도"] = df["시도"].apply(lambda x: x.strip())
df
```

```python
df1 = df.groupby("시도")["등록외국인수"].sum().reset_index()
df1
```

```python
fig, ax = plt.subplots()

plt.rcParams["font.family"] = "KoPubBatang"
ax.bar("시도", "등록외국인수", data=df1)
ax.set_xlabel("시도", size= 15)
ax.set_ylabel("등록외국인수", size = 15)
#ax.set_xticks(range(len(df1)))
#ax.set_xticklabels(df1["시도"], rotation=90)
ax.set_xticks(range(len(df1)), df1["시도"], rotation=90)
plt.show()
```

```python
# bar의 색깔을 다양하게 구성 가능

import matplotlib.pyplot as plt

fig, ax = plt.subplots()

plt.rcParams["font.family"] = "KoPubBatang"

# 색상 리스트 정의 (시도 개수에 맞게 길이 조절 필요)
colors = plt.cm.tab20.colors[:len(df1)]  # 또는 원하는 색상 리스트로 대체 가능

# 컬러맵 정보 참고 --> https://matplotlib.org/stable/users/explain/colors/colormaps.html

ax.bar(df1["시도"], df1["등록외국인수"], color=colors)
ax.set_xlabel("시도", size= 15)
ax.set_ylabel("등록외국인수", size = 15)
ax.set_xticks(range(len(df1)), df1["시도"], rotation=90)
# plt.xticks(rotation=90)
plt.show()
```

```python
# bar의 색깔을 다양하게 구성 가능

import matplotlib.pyplot as plt

fig, ax = plt.subplots()

plt.rcParams["font.family"] = "KoPubBatang"

colors = []

for i in range(len(df1)):
    colors.append("C"+str(i))

ax.bar(df1["시도"], df1["등록외국인수"], color=colors)
ax.set_xlabel("시도", size= 15)
ax.set_ylabel("등록외국인수", size = 15)
plt.xticks(rotation=90)
plt.show()
```

# Pie plot

```python
df = pd.read_csv("pandas_sample#1.csv", encoding='cp949')
df["시도"] = df["시도"].apply(lambda x: x.strip())
df
```

```python
df1 = df[df["년"] == 2024]
df1
```

```python
df2 = df1[df1["시도"] == "서울특별시"]
df2
```

```python
df3 = df2.groupby("시군구")["등록외국인수"].sum().reset_index()
df3
```

```python
fig, ax = plt.subplots()

ax.pie(df3["등록외국인수"], labels=df3["시군구"], autopct="%.1f%%")

plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# 사용자 정의 함수: 실제 값을 포맷팅해서 표시
def real_number(pct, allvals):
    absolute = int(pct / 100. * sum(allvals))
    return f"{absolute:,}"  # 천 단위 쉼표 포함

ax.pie(
    df3["등록외국인수"],
    labels=df3["시군구"],
    autopct=lambda pct: real_number(pct, df3["등록외국인수"])
)

plt.show()
```

```python
fig, ax = plt.subplots()

ax.pie(df3["등록외국인수"], labels=df3["시군구"], autopct="%.1f%%", explode=[0.05]*len(df3))
ax.set_title("서울특별시 등록외국인 현황")

plt.show()
```

```python
# 실습
# 2023년 대전광역시 해당 구별 등록 외국인 현황을 pie 그래프로 나타내어 보시오.
# 2024년 광주광역시 해당 구별 등록 외국인 현황을 pie 그래프로 나타내어 보시오.
```

```python
# 데이터 생성
np.random.seed(0)

x = np.random.rand(100) # 0부터 1까지 무작위 수를 100개 생성
y = np.random.rand(100)
colors = np.random.rand(100)   # 점마다 다른 색상
sizes = 200 * np.random.rand(100)  # 점마다 다른 크기

fig, ax = plt.subplots()

# 산점도
s = ax.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='coolwarm')

# 컬러바 추가
plt.colorbar(s, label='Color Scale')

# 라벨 및 제목
ax.set_xlabel('X 값')
ax.set_ylabel('Y 값')
ax.set_title('컬러와 크기가 있는 산점도')

plt.show()
```

# [Optional] Heatmap

```python
# 지역별 월별 등록외국인수 추이 분석

df = pd.read_csv("pandas_sample#1.csv", encoding="cp949")
df1 = df[df["년"] == 2024]
df_pivot = df1.pivot_table(index=["시도"], columns=["월"], values=["등록외국인수"], aggfunc = "sum")
df_pivot
```

```python
fig, ax = plt.subplots()

im = ax.imshow(df_pivot, cmap="coolwarm")
ax.set_xticks(range(len(df_pivot["등록외국인수"].columns)))
ax.set_yticks(range(len(df_pivot.index)))
ax.set_xticklabels(df_pivot["등록외국인수"].columns)
ax.set_yticklabels(df_pivot.index)
ax.set_xlabel("월")
ax.set_ylabel("지역")
ax.set_title("지역별 월별 등록외국인수 현황")
fig.colorbar(im, ax=ax)
plt.show()
```
