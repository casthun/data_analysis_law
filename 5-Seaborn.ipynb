- Seaborn은 좀 더 복잡한 그래프를 간단한 코드로 구현할 수 있게 해주는 기능을 제공함.

```python
lst = ["대전광역시", "울산광역시", "광주광역시", "대구광역시", "부산광역시", "인천광역시"]

fig, ax = plt.subplots()
plt.rcParams["font.family"] = "KoPubBatang"

# seaborn에서 색상 리스트 얻기 ('pastel', 'tab10', 'husl', 'dark', 'Set2' 등)
# https://seaborn.pydata.org/tutorial/color_palettes.html
# https://colorbrewer2.org

colors = sns.color_palette("muted")

for i, leg in enumerate(lst):
    df2 = df1[df1["시도"]==leg]
    df3 = df2.groupby("월")["등록외국인수"].sum().reset_index()
    sns.lineplot(x="월", y="등록외국인수", data=df3, ax=ax, label=leg, linestyle="--", marker="o", color=colors[i])

ax.set_xticks([x+1 for x in range(12)])

# loc 인자를 이용하여 legend 위치를 미세하게 조정 가능
plt.legend(loc=[0.75, 0.55])

plt.show()
```

# Box plot

```python
df = pd.read_csv("pandas_sample#1.csv", encoding="cp949")
df["시도"] = df["시도"].apply(lambda x: x.strip())
df
```

```python
fig, ax = plt.subplots()

# hue="시도"로 각 막대에 색상 부여, legend=False로 범례 숨김
sns.boxplot(x="시도", y="등록외국인수", hue="시도", data=df, palette="pastel")

plt.xticks(rotation=45)

plt.show()
```

```python
reg_lst = [list(df[df["시도"] == reg]["등록외국인수"]) for reg in df["시도"].unique()]
reg_lst
```

```python
fig, ax = plt.subplots()

reg_lst = []
regions = df["시도"].unique()

for reg in regions:
    df1 = df[df["시도"] == reg]
    reg_temp = list(df1["등록외국인수"])
    reg_lst.append(reg_temp)

# 박스플롯 생성 - patch_artist=True 필수!
box = ax.boxplot(reg_lst, patch_artist=True)

# seaborn에서 색상 리스트 얻기 ('pastel', 'tab10', 'husl', 'dark', 'Set2' 등)
colors = sns.color_palette("pastel", n_colors=len(regions))

# 박스 색 지정
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_edgecolor("gray")

for element in ['whiskers', 'caps', 'medians']:
    for item in box[element]:
        item.set_color("gray")

for flier in box['fliers']:
    flier.set(marker='o', color="grey", alpha=0.5)

# x축 라벨
ax.set_xticks(range(1, len(regions)+1))
ax.set_xticklabels(regions, rotation=45)

plt.title("시도별 등록외국인수 분포")
plt.tight_layout()
plt.show()
```

# Heatmap

```python
df = pd.read_csv("pandas_sample#1.csv", encoding="cp949")
df["시도"] = df["시도"].apply(lambda x: x.strip())
df
```

```python
df1 = df[df["년"] == 2024]
df1
```

```python
df_pivot = df1.pivot_table(index="시도", columns="월", values="등록외국인수", aggfunc="sum")
df_pivot
```

```python
fig, ax = plt.subplots()

sns.heatmap(df_pivot, cmap="crest", ax=ax)

plt.show()
```

```python
# matplotlib으로 heat map 구현하기

df = pd.read_csv("pandas_sample#1.csv", encoding="cp949")
df["시도"] = df["시도"].apply(lambda x: x.strip())
df1 = df[df["년"] == 2024]
df_pivot = df1.pivot_table(index="시도", columns="월", values="등록외국인수", aggfunc="sum")

# 히트맵 그리기
fig, ax = plt.subplots(figsize=(10, 6))  # 그래프 크기 조정

im = ax.imshow(df_pivot, cmap='crest')  # 색상 지정

# 축 설정
ax.set_xticks(np.arange(len(df_pivot.columns)))
ax.set_yticks(np.arange(len(df_pivot.index)))
ax.set_xticklabels(df_pivot.columns)
ax.set_yticklabels(df_pivot.index)

# 값에 대한 컬러바 추가
plt.colorbar(im)

# 제목 추가
ax.set_xlabel("월")
ax.set_ylabel("시도")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("pandas_sample#1.csv", encoding="cp949")
df["시도"] = df["시도"].apply(lambda x: x.strip())
df
```

```python
df1 = df[df["년"] == 2022]
df2 = df[df["년"] == 2023]
df3 = df[df["년"] == 2024]
```

```python
df1g = df1.groupby("월")["등록외국인수"].sum().reset_index()
df2g = df2.groupby("월")["등록외국인수"].sum().reset_index()
df3g = df3.groupby("월")["등록외국인수"].sum().reset_index()
```

```python
fig, ax = plt.subplots(3, 1, figsize=(10, 12))

plt.rcParams["font.family"] = "AppleGothic"

sns.lineplot(x="월", y="등록외국인수", data=df1g, ax = ax[0], color= "C0", linestyle="-.", marker="o", label="2022년")
sns.lineplot(x="월", y="등록외국인수", data=df2g, ax = ax[1], color = "C1", linestyle="--", marker="*", label="2023년")
sns.lineplot(x="월", y="등록외국인수", data=df3g, ax = ax[2], color= "C2", linestyle=":", marker="^", label="2024년")

ax[0].set_xticks(np.arange(1, 13), [x+1 for x in range(12)])
ax[1].set_xticks(np.arange(1, 13), [x+1 for x in range(12)])
ax[2].set_xticks(np.arange(1, 13), [x+1 for x in range(12)])

plt.legend()

plt.show()
```
